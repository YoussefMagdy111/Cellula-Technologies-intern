```python
plt.figure(figsize=(8,5))
plt.hist(data['fare_amount'], bins=50, edgecolor='k')
plt.title("Distribution of Fare Amount")
plt.xlabel("Fare Amount")
plt.ylabel("Frequency")
plt.show()
### insights
#Most fares are concentrated at the lower end (0–20 units)
#The highest frequency is between $5–15, which likely represents typical short intra-city rides
#This indicates that the platform’s core demand is for short, affordable rides
# fee shewayet feres 2orayeba mn el 0 momoken teb2a offers aw demo ride
#revenue may come from frequent short trips, the company might optimize pricing or loyalty programs around this segment
```

```python
sns.lineplot(x="passenger_count", y="fare_amount", data=data)
plt.title("Passenger count vs fare")
plt.xlabel("Passenger count")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x="day", y="fare_amount", data=data)
plt.title("fare per day")
plt.xlabel("Day")
plt.ylabel("Fare")
plt.show()
# all the fares are nearly so the day of the month doesn't make any difference
```

```python
import calendar

data["month_name"] = data["month"].apply(lambda x: calendar.month_name[int(x)])

plt.figure(figsize=(10, 6))
sns.barplot(
    x="month_name", y="fare_amount", data=data, order=list(calendar.month_name)[1:]
)
plt.title("fare per month")
plt.xlabel("Month")
plt.ylabel("Fare")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x="year", y="fare_amount", data=data)
plt.title("fare per year")
plt.xlabel("year")
plt.ylabel("Fare")
plt.show()

# the avg increased after 2012
```

```python
# Boxplot to check outliers
plt.figure(figsize=(6,4))
plt.boxplot(data['fare_amount'], vert=False)
plt.title("Fare Amount Outliers")
plt.show()
##The greatest part of the fare betweeen 5$ and 25$
## we have some outliers 100$+
# we have to do analysis of the biggest fares like it related to some areas like airport or specific time
```

```python
# 5. Distance vs Fare
# ===============================
plt.figure(figsize=(8,5))
plt.scatter(data['distance'], data['fare_amount'], alpha=0.3)
plt.title("Distance vs Fare Amount")
plt.xlabel("Distance (km)")
plt.ylabel("Fare Amount")
plt.show()
## The more distance increase, the more fare amount increase
# we have some outliers in distance 8000+ and the issues is its fare amount is very low from 1$-50$ so it can be entered by error
```

```python
# 6. Passenger Count
# ===============================
plt.figure(figsize=(6,4))
data['passenger_count'].value_counts().sort_index().plot(kind='bar')
plt.title("Passenger Count Distribution")
plt.xlabel("Passengers")
plt.ylabel("Number of Rides")
plt.show()
## most of the rides has one passenger only
## the rides has 6 only very low , it can be that we have few cars their max capacity is 5
#Marketing: Target solo commuters or offer carpooling options for cost savings.
```

```python
plt.figure(figsize=(8,5))
data['weekday'].value_counts().sort_index().plot(kind='bar')
plt.title("Trips by Weekday ")
plt.xlabel("Weekday")
plt.ylabel("Trips")
plt.show()

## Most day that has rides is Friday
## 0 start from sunday
```

```python
plt.figure(figsize=(6,4))
data['Traffic Condition'].value_counts().plot(kind='bar')
plt.title("Traffic Conditions")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['pickup_longitude'], data['pickup_latitude'], alpha=0.2, s=10)
plt.title("Pickup Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
